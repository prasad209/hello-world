```python
DC_for_decision1=DC.copy()
```

```python
DC_for_decision1.head()
```

```python
DC_for_decision1.drop(["GSM"],axis=1,inplace=True)
DC_for_decision1.head()
```

```python
DC_for_decision1.shape
```

```python
DC_for_decision1=DC_for_decision1.where(DC_for_decision1.notnull()).dropna()
DC_for_decision1
```

```python
DC_for_decision1.SEX.unique()
```

```python
DC_for_decision1.isnull().sum()
#now there are no mising values here
```

```python
DC_for_decision1.HAIR.value_counts()
```

```python
independent_variables=["ID","ALIGN","EYE","HAIR","ALIVE","YEAR"]
X=DC_for_decision1[independent_variables]
y=DC_for_decision1["SEX"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=342)
from sklearn.tree import DecisionTreeClassifier
```

```python
model1=DecisionTreeClassifier(criterion="entropy",max_depth=6,random_state=3)
#model1.fit(X_train,y_train)
#Decision tree supports both categorical and numerical data but i see its not working with categorical data, so i
#googled it and as per many articles sklearn is not supporting it currently so going for one-hot encoding , please feel free to correct me here...
#we cant fit this model directly because we have categorical data, for it to fit , data must be numerical, lets do some one-hot encoding
one_hot_encoded_data=pd.get_dummies(DC_for_decision1,columns=["ID","ALIGN","EYE","HAIR","ALIVE","SEX"])
#one_hot_encoded_data is complete df and we get to choose which columns we need one hot encode , only thise columns will be affected
#and we dont need to join them with remaining already numeric columns
```

**How to do One-Hot encoding??** 

```python
one_hot_encoded_data
```

```python
one_hot_encoded_data.columns
```

```python
independent_variables=['ID_Identity Unknown', 'ID_Public Identity', 'ID_Secret Identity',
       'ALIGN_Bad Characters', 'ALIGN_Good Characters',
       'ALIGN_Neutral Characters', 'ALIGN_Reformed Criminals',
       'EYE_Amber Eyes', 'EYE_Black Eyes', 'EYE_Blue Eyes', 'EYE_Brown Eyes',
       'EYE_Gold Eyes', 'EYE_Green Eyes', 'EYE_Grey Eyes', 'EYE_Hazel Eyes',
       'EYE_Orange Eyes', 'EYE_Photocellular Eyes', 'EYE_Pink Eyes',
       'EYE_Purple Eyes', 'EYE_Red Eyes', 'EYE_Violet Eyes', 'EYE_White Eyes',
       'EYE_Yellow Eyes', 'HAIR_Black Hair', 'HAIR_Blond Hair',
       'HAIR_Blue Hair', 'HAIR_Brown Hair', 'HAIR_Gold Hair',
       'HAIR_Green Hair', 'HAIR_Grey Hair', 'HAIR_Orange Hair',
       'HAIR_Pink Hair', 'HAIR_Platinum Blond Hair', 'HAIR_Purple Hair',
       'HAIR_Red Hair', 'HAIR_Reddish Brown Hair',
       'HAIR_Strawberry Blond Hair', 'HAIR_Violet Hair', 'HAIR_White Hair',
       'ALIVE_Deceased Characters', 'ALIVE_Living Characters','YEAR']


dependent_variables=['SEX_Female Characters', 'SEX_Genderless Characters',
       'SEX_Male Characters']
X=one_hot_encoded_data[independent_variables]
y=one_hot_encoded_data[dependent_variables]
```

```python
X_train_enc,X_test_enc,y_train_enc,y_test_enc=train_test_split(X,y,test_size=0.3,random_state=342)
model1.fit(X_train_enc,y_train_enc)
decision_tree_preds=model1.predict(X_test_enc)
decision_tree_train_preds=model1.predict(X_train_enc)

decision_tree_preds
```
